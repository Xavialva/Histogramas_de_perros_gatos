## 1. Traza un histograma del porcentaje de poseedores de perros.

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,6))
plt.hist(df['Percentage of Dog Owners'],bins= 10,edgecolor='k',alpha=0.80,label='Dogs Owners')
plt.xlim(0,60)
plt.xlabel('Porcentaje de dueños de perros')
plt.ylabel('Cantidad de perros')
plt.legend(loc='upper right')


```

## 2. Traza un histograma del porcentaje de poseedores de gatos.

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,6))
plt.hist(df['Percentage of Cat Owners'],bins= 10,edgecolor='k', color= 'orange',alpha=0.60, label='Cats')
plt.xlim(0,60)
plt.legend(loc='upper right')
plt.xlabel('Porcentaje de dueños de gatos')
plt.ylabel('Cantidad de gatos')
```

## 3a. ¿Qué podemos ver al comparar estos dos histogramas? ¿Qué información nos da?

Nos indica que existen más cantidad de dueños de gatos que de perros

## 3b. Opcional/Extra: traza estos dos histogramas en el mismo plot. A continuación, un ejemplo de cómo debería lucir:

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,6))
plt.hist(df['Percentage of Dog Owners'],bins= 7,edgecolor='k',alpha=0.80, label='Dogs Owners')
plt.hist(df['Percentage of Cat Owners'],bins= 15,edgecolor='k', color= 'orange',alpha=0.751, label='Cats')
plt.xlim(10,51)
plt.ylim(0,15)
plt.legend(loc='upper right')
plt.xlabel('Porcentaje de dueños de perros y gatos')
plt.ylabel('Cantidad de gatos y perros')

plt.show()
```

## 4. Crea dos diagramas de caja en el mismo plot: uno con el número medio de gatos por hogar y el otro, del número medio de perros. Más abajo, un ejemplo de cómo podría lucir:

```python

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot([df['Mean Number of Cats'],df['Mean Number of Dogs per household']],
           notch = True,
           labels=['Cats', 'Dogs'],
           widths = 0.7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          );

boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('red')

plt.ylabel('Mean Number of Animals per State', fontsize = 20);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
```

## 5. ¿Qué podemos ver al comparar estos dos diagramas de caja? ¿Qué información nos da?
Podemos observar que por cada hogar se tiene en promedio 2 gatos, mientras que en perros es de un 1.6. La máxima cantidad de gatos por hogar es de un 2.6 y la mínima es de 1.7, mientras que en los perros la máxima por familia en cada uno de los estados es de 2.2 y la mínima es de 1 perro por casa.
